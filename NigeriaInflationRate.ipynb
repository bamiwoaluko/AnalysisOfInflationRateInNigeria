```python
import pandas as pd
import matplotlib.pyplot as plt
# reading the csv file
df = pd.read_csv('NigeriaInflationRate2.csv')
df['date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format="mixed") 
df.sort_values('date', inplace=True)

# plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['Inflation_Rate'], marker='o', linestyle='-', color='b', label='Inflation Rate')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Inflation Rate')
plt.title('Inflation Rate Over Time in Nigeria')
plt.legend()

# Show the plot
plt.show()
```
